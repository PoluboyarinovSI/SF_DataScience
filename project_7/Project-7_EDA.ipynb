# 0. Постановка задачи

В данном проекте решается бизнес-задача в области безопасности банковского сектора. Необходимо создать комплекс мер для обнаружения и предотвращения мошеннических транзакций. 
Платежное мошенничество или мошенничество с платежами дистанционного банковского обслуживания (ДБО) - один из видов банковского мошенничества с банковскими картами, когда мошенник осуществляет платежи по карте без ведома владельца.
Выявление мошенничества с платежами - это комплекс мер, направленных на обнаружение различных рисков мошенничествао с платежами, которые могут нанести ущерб. Одним из наиболее эффективных инструментов для этого является машинное обучение. Одной из основных  трудностей в области обнаружения мошеннических транзакций является низкая доля мошеннических транзакций от общего числа транзакций (как правило менее 1%)

**Информация об исходных данных** В качестве исходных данных использован смоделированный набор данных о транзакциях по кредитным картам, содержащий законные и мошеннические транзакции за период с 1 января 2019 года по 31 декабря 2020 года. Набор данных охватывает кредитные карты 1000 клиентов, совершающих транзакции с пулом из 800 продавцов. Данные были взяти из Kaggle по ссылке: https://www.kaggle.com/datasets/kartik2112/fraud-detection?select=fraudTrain.csv

**Бизнес-задача:** Создать механизм обнаружения мошеннических транзакций .

**Техническая задача:** построить модель классификации на основе набора данных с транзакциями по кредитным картам для обнаружения мошеннических транзакций.

**Основные цели проекта:**
1. Провести предобработку набора данных.
2. Провести разведывательный анализ данных и выявить основные закономерности.
3. Провести подготовку признаков для обучения/создать новые признаки. 
4. Построить несколько моделей машинного обучения, решающих задачу классификации транзакций по банковским картам.
5. Спроектировать процесс классификации транзакций по банковким картам с целью обнаружения мошеннических транзакций.

## Выбор метрики

Точность классификации - это показатель, который обобщает производительность модели классификации как количество правильных предсказаний, деленное на общее количество прогнозов.
Accuracy = Correct Predictions / Total Predictions

В случае несбалансированности предсказываемых классов в данных достижение высокго показателя Accuracy является обычной ситуацией. Поэтому показатель Accuracy не применим в несбалансированном наборе данных.

В задачах, связанных с обнаружением высокого риска, метрика полноты (Recall) является более важным показателем оценки, чем точность. В случае обнаружения мошеннических операций с кредитными картами стоит задача максимально избежать ложных срабатываний. Мошеннические операции наносят большой ущерб, поэтому необходимо принимать соответствующие меры для их предотвращения. Ложноотрицательный случай в данном контексте означает, что мошенническая транзакция оценивается как обычная, что наносит ущерб. Поэтому ложные срабатывания не так важны, как предотвращение мошенничества.

Полнота рассчитывается как отношение числа Positive выборок, корректно классифицированных как Positive, к общему количеству Positive семлов. Полнота измеряет способность модели обнаруживать выборки, относящиеся к классу Positive. Чем выше метрика полноты, тем больше Positive семплов было обнаружено моделью.

# 2. Data pre-processing

```python
# Feautures type transformation
train_data['trans_date_trans_time'] = pd.to_datetime(train_data['trans_date_trans_time'])
train_data['dob'] = pd.to_datetime(train_data['dob'])
train_data['zip'] = train_data['zip'].astype(dtype='string')
```

```python
def add_datetime_features(data):
    """
    Function for adding date features such as:
    transaction month, day of week, hour and age of cardholder

    Args:
        data (DataFrame): Original dataset

    Returns:
        data (DataFrame): Modified dataset
    """
    data['trans_month_of_year'] = data['trans_date_trans_time'].dt.month
    data['trans_day_of_week'] = data['trans_date_trans_time'].dt.day_of_week
    data['trans_hour'] = data['trans_date_trans_time'].dt.hour
    data['age_of_cardholder'] = np.round((data['trans_date_trans_time'] - data['dob']) / np.timedelta64(1, 'Y'))
    data = data.drop(['trans_date_trans_time'], axis=1)
    return data
```

```python
# Adding date features
pre_processed_data = add_datetime_features(train_data)
```

```python
def add_age_group(age):
    """
    Function for adding age group of cardholders

    Args:
        age (float): Age of cardholder

    Returns:
        string: Cardholders age group
    """
    if age <= 30:
        return '< 30'
    elif age > 30 and age <= 45:
        return '30-45'
    elif age > 45 and age <= 60:
        return '45-60'
    else:
        return '> 60'
```

```python
# Adding of cardholders age group
pre_processed_data['age_group'] = pre_processed_data['age_of_cardholder'].apply(add_age_group)
```

```python
# Age groupe of cardholder feature analysis
overal_age_grouped = pre_processed_data[['age_group', 'trans_num']].groupby(['age_group']).count().reset_index().rename(columns={'trans_num':'trans_count'})
fraud_age_grouped = pre_processed_data[['age_group', 'trans_num', 'is_fraud']].groupby(['age_group', 'is_fraud']).count().reset_index().rename(columns={'trans_num':'fraud trans_count'})
merged_age_grouped = fraud_age_grouped.merge(overal_age_grouped, how='inner', on='age_group')
merged_age_grouped['percentage'] = merged_age_grouped['fraud trans_count'] / merged_age_grouped['trans_count'] * 100
merged_age_grouped
```

## 4.2 Feature Creation

```python
def add_high_risk_job(data, high_risk_list):
    """Function for adding high risk job mark

    Args:
        data (DataFrame): Pre-processed dataframe
        high_risk_list (list): List of high risk jobs titles

    Returns:
        DataFrame: Processed dataframe with high risk job mark feature
    """
    data['high_risk_job'] = data['job'].apply(lambda x: 1 if x in high_risk_list else 0)
    data['job'] = data['job'].apply(lambda x: x if x in high_risk_list else 'Other')
    return data

def add_high_risk_merchant(data, high_risk_list):
    """Function for adding high risk merchant mark

    Args:
        data (DataFrame): Pre-processed dataframe
        high_risk_list (list): List of high risk merchant titles

    Returns:
        DataFrame: Processed dataframe with high risk merchant mark feature
    """
    data['high_risk_merchant'] = data['merchant'].apply(lambda x: 1 if x in high_risk_list else 0)
    data['merchant'] = data['merchant'].apply(lambda x: x if x in high_risk_list else 'Other')
    return data

def add_high_risk_city(data, high_risk_list):
    """Function for adding high risk city mark

    Args:
        data (DataFrame): Pre-processed dataframe
        high_risk_list (list): List of high risk city titles

    Returns:
        DataFrame: Processed dataframe with high risk city mark feature
    """
    data['high_risk_city'] = data['city'].apply(lambda x: 1 if x in high_risk_list else 0)
    data['city'] = data['city'].apply(lambda x: x if x in high_risk_list else 'Other')
    return data

def add_high_risk_zip(data, high_risk_list):
    """Function for adding high risk zip mark

    Args:
        data (DataFrame): Pre-processed dataframe
        high_risk_list (list): List of high risk zip titles

    Returns:
        DataFrame: Processed dataframe with high risk zip mark feature
    """
    data['high_risk_zip'] = data['zip'].apply(lambda x: 1 if x in high_risk_list else 0)
    data['zip'] = data['zip'].apply(lambda x: x if x in high_risk_list else 'Other')
    return data
```

```python
# Features creating
pre_processed_data = add_high_risk_job(pre_processed_data, high_risk_job)
pre_processed_data = add_high_risk_merchant(pre_processed_data, high_risk_merchant)
pre_processed_data = add_high_risk_city(pre_processed_data, high_risk_cities)
pre_processed_data = add_high_risk_zip(pre_processed_data, high_risk_zip)
```
